```python
from sklearn.metrics import mean_squared_error

def rmse(y_true, y_pred):
    return np.sqrt(mean_squared_error(y_true, y_pred))
```

```python
# 50% training and 50% testing
X_train_50, X_test_50, y_train_50, y_test_50 = train_test_split(nyc.Date.values.reshape(-1, 1), nyc.Temperature.values, test_size=0.5, random_state=11)

linear_regression_50 = LinearRegression()
linear_regression_50.fit(X=X_train_50, y=y_train_50)
predicted_50 = linear_regression_50.predict(X_test_50)

for p, e in zip(predicted_50, y_test_50):
    print(f"predicted: {p:.2f}, expected: {e:.2f}")

rmse_50 = rmse(y_test_50, predicted_50)
print(f"RMSE: {rmse_50:.2f}")
```

```python
# 80% training and 20% testing
X_train_80, X_test_80, y_train_80, y_test_80 = train_test_split(nyc.Date.values.reshape(-1, 1), nyc.Temperature.values, test_size=0.2, random_state=11)

linear_regression_80 = LinearRegression()
linear_regression_80.fit(X=X_train_80, y=y_train_80)
predicted_80 = linear_regression_80.predict(X_test_80)

for p, e in zip(predicted_80, y_test_80):
    print(f"predicted: {p:.2f}, expected: {e:.2f}")

rmse_80 = rmse(y_test_80, predicted_80)
print(f"RMSE: {rmse_80:.2f}")
```

```python
# 90% training and 10% testing
X_train_90, X_test_90, y_train_90, y_test_90 = train_test_split(nyc.Date.values.reshape(-1, 1), nyc.Temperature.values, test_size=0.1, random_state=11)

linear_regression_90 = LinearRegression()
linear_regression_90.fit(X=X_train_90, y=y_train_90)
predicted_90 = linear_regression_90.predict(X_test_90)

for p, e in zip(predicted_90, y_test_90):
    print(f"predicted: {p:.2f}, expected: {e:.2f}")

rmse_90 = rmse(y_test_90, predicted_90)
print(f"RMSE: {rmse_90:.2f}")
```

```python
# 75% training and 25% testing
X_train_75, X_test_75, y_train_75, y_test_75 = train_test_split(nyc.Date.values.reshape(-1, 1), nyc.Temperature.values, test_size=0.25, random_state=11)

linear_regression_75 = LinearRegression()
linear_regression_75.fit(X=X_train_75, y=y_train_75)
predicted_75 = linear_regression_75.predict(X_test_75)

for p, e in zip(predicted_75, y_test_75):
    print(f"predicted: {p:.2f}, expected: {e:.2f}")

rmse_75 = rmse(y_test_75, predicted_75)
print(f"RMSE: {rmse_75:.2f}")
```
